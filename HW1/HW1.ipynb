```python
import sys
sys.path.append('/content/drive/My Drive')

from graph_ml import GraphML

import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.karate_club_graph()
```

```python
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title("Zachary's Karate Club Graph")
plt.show()
```

```python
graph_ml = GraphML(edge_list=G.edges())
```

```python
print("BFS:", graph_ml.bfs(0))
```

```python
print("DFS:", graph_ml.dfs(0))
```

```python
print("Shortest path (4 -> 14):", graph_ml.shortest_path(4, 14))
```

```python
print("All paths of length 3 (between 2 and 29):", graph_ml.all_paths_length(2, 29, 3))
```

```python
print("Cycles:", graph_ml.find_cycles())
```

```python
print("Degree matrix:\n", graph_ml.degree_matrix())
```

```python
local_overlap = graph_ml.local_overlap_measures()
print("Local overlap measures:")
for (u, v), value in local_overlap.items():
    print(f"({u}, {v}): {value:.4f}")
```

```python
print("Laplacian matrix:\n", graph_ml.laplacian_matrix())
```

```python
eigenvalues, eigenvectors = graph_ml.adjacency_eigenvalues_vectors()

print("Eigenvalues of the adjacency matrix:")
print(eigenvalues)

print("\nEigenvectors of the adjacency matrix:")
print(eigenvectors)
```

```python
k = 3
clusters = graph_ml.spectral_clustering(k=k)

plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green']
node_colors = [colors[label] for label in clusters]

nx.draw(G, with_labels=True, node_color=node_colors, edge_color='gray', cmap=plt.cm.rainbow)
plt.title(f"Spectral Clustering (k={k})")
plt.show()

print(f"Labels (k={k}):", clusters)
```
